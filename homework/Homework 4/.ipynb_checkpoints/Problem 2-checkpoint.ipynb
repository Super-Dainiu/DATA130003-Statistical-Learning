```r
prob<-function(x,b){
    return(exp(x%*%b)/(1+exp(x%*%b)))
}
l<-function(x,y,b){
    return(t(y)%*%x%*%b-sum(log(1+exp(x%*%b))))
}
```

```r
NR<-function(N){
    train.X<-matrix(c(rep(1,N),rnorm(2*N)),nrow=N)
    b<-c(0.5,1.2,-1)
    train.Y<-prob(train.X,b)>runif(N)
    train.b<-rep(0.5,3)
    hist<-train.b
    for (i in 1:200){
        g<-t(train.Y-prob(train.X,b))%*%train.X
        W<-diag(as.vector(prob(train.X,b)*(1-prob(train.X,b))))
        h<--t(train.X)%*%W%*%train.X
        k<-1
#         while ((l(train.X,train.Y,train.b+k*solve(h)%*%t(g))-l(train.X,train.Y,train.b))<k*sum((solve(h)%*%t(g))^2)&(k>1e-3)) k<-k/2
        if (max(abs(solve(h)%*%t(g))<1e-5)) break
        train.b<-train.b+k*solve(h)%*%t(g)
        hist<-cbind(hist,train.b)
    }
    return(hist)
}
b<-c(0.5,1.2,-1)
NR(200)-b
NR(500)-b
NR(800)-b
NR(1000)-b
```
